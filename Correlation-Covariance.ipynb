```python
# Importing numpy library
import numpy as np
```

```python
# Covariance function
def covariance_func(a,b):
    a=np.array(a)
    b=np.array(b)
    mean_a=0
    for i in a:
        mean_a=mean_a+i
    mean_a=mean_a/len(a)

    mean_b=0
    for i in b:
        mean_b=mean_b+i
    mean_b=mean_b/len(b)


    
    new=(a-mean_a)*(b-mean_b)
    covariance=0
    for i in new:
        covariance+=i
    covariance=covariance/len(new)
    return covariance
```

```python
# Defining lists
a=[1,2,3,4,5]
b=[10,20,30,40,50]
c=[5000,4000,3000,2000,1000]
d=[500,100,400,200,300]
```

```python
#observing covariance values
print('list a and b with positive correlation, covariance:{}'.format(covariance_func(a,b)))
print('list a and c with negative correlation, covariance:{}'.format(covariance_func(a,c)))
print('list a and d with zero correlation, covariance:{}'.format(covariance_func(a,d)))
```

```python
# Variance function
def variance_func(a):
    a=np.array(a)
    mean_a=0
    for i in a:
        mean_a=mean_a+i
    mean_a=mean_a/len(a)

    new=(a-mean_a)**2

    variance=0
    for i in new:
        variance+=i

    variance=variance/len(a)

    return variance
    
```

```python
# Correlation function
def correlation_func(a,b):
    covariance=covariance_func(a,b)
    variance_a=variance_func(a)
    variance_b=variance_func(b)
    correlation=covariance/((variance_a**0.5)*(variance_b**0.5))
    return correlation
```

```python
print('list a and b with positive correlation, correlation:{:.2f}'.format(correlation_func(a,b)))
print('list a and c with negative correlation, correlation:{:.2f}'.format(correlation_func(a,c)))
print('list a and d with zero correlation, correlation:{:.2f}'.format(correlation_func(a,d)))
```
