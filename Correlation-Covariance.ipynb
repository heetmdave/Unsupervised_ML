```python
# Importing numpy library
import numpy as np
```

```python
# Covariance function
def covariance_func(a,b):
    a=np.array(a)
    b=np.array(b)
    mean_a=np.mean(a)
    mean_b=np.mean(b)
    new=(a-mean_a)*(b-mean_b)
    covariance=0
    for i in new:
        covariance+=i
    covariance=covariance/len(new)
    return covariance
```

```python
# Defining lists
a=[1,2,3,4,5]
b=[10,20,30,40,50]
c=[5000,4000,3000,2000,1000]
d=[500,100,400,200,300]
```

```python
#observing covariance values
print('list a and b with positive correlation, covariance:{}'.format(covariance_func(a,b)))
print('list a and c with negative correlation, covariance:{}'.format(covariance_func(a,c)))
print('list a and d with zero correlation, covariance:{}'.format(covariance_func(a,d)))
```
