```python
#importing necessary libraries
import numpy as np
```

```python
#define two features
f1=np.array([1,2,3,4,5])
f2=np.array([1,2,3,4,5])
```

```python
#mean centering the data
f1=f1 - np.mean(f1)
f2=f2 - np.mean(f2)
```

```python
#create a 2D array where each row is a feature
arr=np.array([f1,f2])
arr
```

```python
#calculate covariance matrix
cov=np.cov(arr)
cov
```

```python
#calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov)
eigenvalues, eigenvectors
```

```python
#getting index of sorted eigenvalues in descending order
index=np.argsort(eigenvalues)[::-1]
```

```python
#selecting the eigenvector with the highest eigenvalue
new_bases=eigenvectors[:,0]
```

```python
arr.T
```

```python
#projecting the data onto the new basis
new_feature=np.dot(arr.T,new_bases)
new_feature
```
